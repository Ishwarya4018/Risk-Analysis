```python
bins = [0, 100000, 300000, float('inf')]
labels=['Low', "Med", "High"]
df['Income Band']=pd.cut(df["Estimated Income"],bins=bins, labels=labels,right=False)
df['Income Band'].value_counts().plot(kind="bar",color="yellow")
```

The distribution of unique categorical columns

```python
columns_to_check = ['GenderId', 'Amount of Credit Cards','BRId',"IAId","Nationality","Fee Structure","Loyalty Classification","Properties Owned","Income Band"]  
for column in columns_to_check:

        print(f"\nValue counts for column: '{column}'")
        print(df[column].value_counts())
        print('-' * 40)
```

```python
import seaborn as sns
for i, predictor in enumerate(df):
    plt.figure(i)
    sns.countplot(data=df,x=predictor,hue="GenderId")
```

```python
for i, predictor in enumerate(df):
    plt.figure(i)
    sns.countplot(data=df,x=predictor,hue="Nationality")
```

Numerical Analysis

```python
numerical_columns= ['Estimated Income',"Superannuation Savings","Credit Card Balance","Bank Loans","Bank Deposits","Checking Accounts","Foreign Currency Account","Saving Accounts"]
plt.figure(figsize=(8,6))
for i,col in enumerate(numerical_columns):

    sns.histplot(df[col],kde=True)
    plt.title(col)
    plt.show()
```

Heatmaps

```python
numerical_columns= ['Estimated Income',"Superannuation Savings","Credit Card Balance","Bank Loans","Bank Deposits","Checking Accounts","Foreign Currency Account","Saving Accounts"]
correlation_matrix=df[numerical_columns].corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation_matrix,annot=True,cmap="crest",fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

INSIGHTS
1. A majority of clients fall under the High Fee Structure contributing over 44 crores which is more than 75% of the total fee revenue.
2. Most clients with high fee are within 24-40 age group.
3. Clients with igher fee structure have higher deposits and higher checking and saving accounts.
4. Most clients hold multiple credit cards and credit card balances.
5. Positive correlation betwween "Estimated Income" and "Superannuation Savings".
6. Negative or weak correlation between "Properties owned" and "Risk Weighting".
